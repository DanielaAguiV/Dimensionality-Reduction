```python
my_imagen = 'D:/Documentos/Especializacion/Ml_2/Task_1/Dimensionality-Reduction/Lab1_Task2/pictureFace/DanielaAguiarV.jpg'
my_face_edited = picture.Picture(my_imagen).image_asarray()
my_face_edited = np.array(my_face_edited)
my_svd = SVD_np.SVD(n_components= 256)
my_svd.fit(my_face_edited)
my_face_edited_svd =my_svd.transform(my_face_edited)
picture.PictureOperations().image_fromarray(my_face_edited_svd)
```

```python
from sklearn.decomposition import TruncatedSVD
my_imagen = 'D:/Documentos/Especializacion/Ml_2/Task_1/Dimensionality-Reduction/Lab1_Task2/pictureFace/DanielaAguiarV.jpg'
my_face_edited = picture.Picture(my_imagen).image_asarray()
my_face_edited = np.array(my_face_edited)
svd_sk = TruncatedSVD(n_components = 256 )
svd_sk.fit(my_face_edited)
imagen_from_sk = svd_sk.transform(my_face_edited)
picture.PictureOperations().image_fromarray(imagen_from_sk)
```
